# Aula 11. Módulos, Pacotes e Bibliotecas
## Módulos
Um módulo é simplesmente um arquivo de código que contém funções, classes e variáveis reutilizáveis. 

Os módulos são úteis porque permitem que você divida seu código em partes menores e mais gerenciáveis, e também permitem que você reutilize código em vários projetos.

## Pacotes
Um pacote em Python é uma forma de organizar e agrupar módulos de código. Cada módulo é um arquivo .py que contém funções, classes, variáveis e outros elementos de código. 

Agrupar esses módulos em pacotes ajuda a manter o código organizado e fácil de entender. Além disso, os pacotes permitem a reutilização do código, já que módulos dentro de um pacote podem ser importados e utilizados em outros programas.

## Bibliotecas
Uma biblioteca é um conjunto de módulos relacionados que fornecem recursos adicionais para um determinado propósito.

Para importar módulos, pacotes e bibliotecas utilizamos a keyword **import**

```python
import math # Importa o módulo math
```

```python
math.sqrt(9) # Função sqrt -> retorna a raiz quadrada de um número
```

```python
from math import sqrt # Do módulo math importe a função sqrt (raiz quadrada)

sqrt(9)
```

## Pip
Pip (acrônimo para "Package Installer for Python") é um gerenciador de pacotes para o Python. 

Ele permite que você instale bibliotecas e módulos Python de forma fácil e rápida. 

Com o pip, você pode procurar, baixar e instalar pacotes Python a partir do Python Package Index (PyPI), que é uma coleção centralizada de pacotes Python.

O pip é uma ferramenta importante para qualquer pessoa que trabalha com Python, pois permite que você aproveite a riqueza de bibliotecas e módulos que já foram desenvolvidos por outras pessoas, o que pode poupar muito tempo e esforço na resolução de tarefas comuns

## Biblioteca requests

A biblioteca requests é uma biblioteca que fornece uma interface de alto nível para fazer requisições HTTP. 

Ela permite enviar solicitações HTTP (como GET, POST, PUT, DELETE, etc.) a servidores web e ler a resposta retornada, incluindo cabeçalhos e conteúdo. 

A biblioteca requests facilita o processo de requisições HTTP, tornando-o mais fácil de ler,

Aqui está um exemplo básico da biblioteca Requests em Python para realizar uma requisição HTTP GET a um URL:

```python
import requests

url = "https://pt.wikipedia.org/wiki/Python"

response = requests.get(url)

print(response.text)
```
